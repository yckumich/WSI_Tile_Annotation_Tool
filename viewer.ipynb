## WSI Tile Annotation Tool

Mark diagnostically important regions on a whole-slide image, tile by tile. Run the cell below to launch the viewer.

**Starting a new slide**
1. In **Choose file**, pick a `.svs` file from `slides/` and enter your name in **Annotator ID**.
2. Click **Load**. The thumbnail (left) shows the whole slide; the square panel (right) shows a zoomed-in view.
3. Navigate with the zoom slider, by clicking/dragging on the thumbnail, or with the pan sliders.
4. Click or drag on the zoomed view to select tiles (drag paints multiple tiles at once); click a selected tile again to deselect it.
5. Adjacent selected tiles automatically form a **region** — no separate step to create one. Each region gets its own color and its own note box.
6. Regions merge and split automatically as you select/deselect tiles: selecting a tile that touches two different regions joins them into one; deselecting the one tile connecting two parts of a region splits it back into two. Colors and notes carry over sensibly when this happens.
7. Under the thumbnail, each region has a color swatch and a text box — write your reasoning for that specific region there. Click a region's swatch to jump the viewport straight to it.
8. **Reset** clears all current regions and notes (it does not change your zoom/pan position).
9. When you're done, click **Complete** to export your regions to `annotations/` as a timestamped JSON file.

**Resuming a previous session**
1. In **Choose file**, pick a previously exported JSON file from `annotations/` instead of a `.svs` file.
2. Click **Load** — this reopens the original slide and restores your prior regions, colors, and notes exactly as you left them.
3. Continue selecting/deselecting tiles and editing notes, then click **Complete** again to save a new export (your earlier export is never overwritten).
